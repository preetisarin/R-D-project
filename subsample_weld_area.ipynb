```python
file=['subsample/L_I1.wav', 
    'subsample/L_II1.wav',
    'subsample/L_III.wav',
    'subsample/M_I2.wav',
    'subsample/M_II2.wav',
    'subsample/M_IV1.wav',
    'subsample/M_II3.wav',
    'subsample/M_IV2.wav',
    'subsample/M_II4.wav',
    'subsample/M_IV3.wav',
    'subsample/M_V.wav']
```

```python
def energy(signal):
    """Computes signal energy of frame"""
    return np.sum(signal ** 2) / np.float64(len(signal))
```

```python
frame_size=1048
hop_length=256
```

```python
combined=[]
for f in file:
    audio_segment = AudioSegment.from_file(f)
    record=[audio_segment.frame_rate,len(audio_segment),audio_segment.frame_count(),audio_segment.dBFS]
    y,sr=l.load(f)
    record.append(energy(y))
            
    zcr=l.feature.zero_crossing_rate(y=y,frame_length=frame_size,hop_length=hop_length)[0]
    record.append(np.mean(zcr))
            
    rms=l.feature.rms(y=y,frame_length=frame_size,hop_length=hop_length)[0]
    record.append(np.mean(rms))
    
    sc=l.feature.spectral_centroid(y=y, sr=sr,n_fft=frame_size,hop_length=hop_length)[0]
    record.append(np.mean(sc))

    sb=l.feature.spectral_bandwidth(y=y, sr=sr,n_fft=frame_size,hop_length=hop_length)[0] 
    record.append(np.mean(sb))

    mfcc=l.feature.mfcc(y=y,sr=sr)
    record.append(np.mean(mfcc))

    max_frequency=max(y)
    # print(max_frequency)
    record.append(max_frequency)

    min_frequency=min(y)
    # print(min_frequency)
    record.append(min_frequency)
    record.insert(0,f)
    combined.append(record)

```

```python
col_names=['File_name','Frame_rate','Duration (ms)','Frame_count','Intensity',
            'Energy','Zero_crossing_rate','Root Mean Square','Spectral Centroid','Spectral Bandwidth', 'MFCC','Max_frequency','Min_frequency']
with open('CSV/feature_subsample.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(col_names)
    writer.writerows(combined)
```
